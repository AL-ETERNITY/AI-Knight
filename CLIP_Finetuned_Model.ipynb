```python
# Import necessary libraries
import os
from transformers import AutoImageProcessor, AutoModelForImageClassification
from PIL import Image
import torch
import pandas as pd
```

the model is finetned of google/vit-base-patch16-224-in21k available on hugingface

```python
# Load the model and processor
processor = AutoImageProcessor.from_pretrained("dima806/ai_vs_real_image_detection")
model = AutoModelForImageClassification.from_pretrained("dima806/ai_vs_real_image_detection")
```

```python
# Define the dataset path
dataset_path = "/kaggle/input/ai-knight/dataset/train"
fake_folder = os.path.join(dataset_path, "fake")
real_folder = os.path.join(dataset_path, "real")
# Define the test dataset path
test_folder = "/kaggle/input/ai-knight/dataset/test"
```

```python
# Get all image file paths from the test folder
test_file_paths = [os.path.join(test_folder, f) for f in os.listdir(test_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

# Initialize a list to store predictions
predicted_classes = []

# Set model to evaluation mode
model.eval()

# Loop through each test image
for i, image_path in enumerate(test_file_paths):
    # Load and preprocess the image
    image = Image.open(image_path).convert("RGB")
    inputs = processor(image, return_tensors="pt")
    
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Predict the class
    probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
    prediction = torch.argmax(probs, dim=-1).item()
    predicted_classes.append(prediction)
    
    # Print progress
    print(f"Processed {i + 1}/{len(test_file_paths)} images.")
```

```python
# Create a DataFrame for the predictions
output_df = pd.DataFrame({
    'ID': [os.path.basename(f) for f in test_file_paths],  # Extract filenames only
    'ImageType': predicted_classes  # Predictions (0 or 1)
})

# Save the predictions to a CSV file
output_csv_path = '/kaggle/working/test_predictions.csv'
output_df.to_csv(output_csv_path, index=False)
print(f"Predictions saved to {output_csv_path}.")
```

```python
# Path to the original predictions CSV file
original_csv_path = '/kaggle/working/test_predictions.csv'

# Read the CSV file
predictions_df = pd.read_csv(original_csv_path)

# Extract numeric part from the 'ID' column
predictions_df['ID'] = predictions_df['ID'].str.extract(r'(\d+)').astype(int)

# Sort the DataFrame by the numeric 'ID' column
sorted_predictions_df = predictions_df.sort_values(by='ID').reset_index(drop=True)

# Save the sorted DataFrame to a new CSV file
sorted_csv_path = '/kaggle/working/sorted_test_predictions2.csv'
sorted_predictions_df.to_csv(sorted_csv_path, index=False)

print(f"Sorted predictions saved to {sorted_csv_path}.")
```
